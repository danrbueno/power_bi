# Titles and genres dataset

*   This Python script list all the titles from a [public Netflix dataset](https://www.kaggle.com/datasets/victorsoeiro/netflix-tv-shows-and-movies) and creates a new dataset with each title and country in a separated row.

```python
import pandas as pd
```

Read the CSV file that contains the titles from Netflix dataset.

```python
titles = pd.read_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/titles.csv")
titles
```

Read the CSV file that contains the list of all countries.

```python
countries = pd.read_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/countries.csv", index_col="code")
countries
```

Every row can have more than one genre.

So, it's necessary to separate the genres listed in every row.

This column will be treated as list of strings.

```python
production_countries = titles.production_countries
production_countries
```

Read all the genres listed and create a new dataset relationing all titles and genres.

```python
df = pd.DataFrame()
#print(unique_genres.where(unique_genres.genre == 'documentation').head(1).id)
for title in titles.itertuples():
  for country_code in eval(title.production_countries):
    
    if country_code in countries.index:
      country_id = str(countries.loc[country_code].id)
    else:
      country_id = "999"
    
    row = {'title_id': title.id,
          'country_id': country_id}
    df = df.append(row, ignore_index=True)
  
print(df)
df.to_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/titles_countries.csv", encoding = "utf-8", index=False)
```
