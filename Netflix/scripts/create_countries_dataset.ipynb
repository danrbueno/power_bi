# Countries dataset

*   This Python script list all the countries from a Wikipedia page and generates a dataset with the codes and names of the countries.

```python
import pandas as pd
```

Get the country names from a wiki table.

```python
from pickle import FALSE
wiki = pd.read_html("https://en.wikipedia.org/wiki/ISO_3166-1_alpha-2", keep_default_na = False)
countries = wiki[2]
countries
```

Generates another dataset with the list of countries.



```python
df = pd.DataFrame()
i = 1

for country in countries.itertuples():
  row = {'id': str(i),
         'code': country.Code,
         'name': country[2]}
  df = df.append(row, ignore_index=True)
  i = i+1

not_registered = {'id': "999",
                  'code': "99",
                  'name': "Not Registered"}
df = df.append(not_registered, ignore_index=True)

print(df)

df.to_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/countries.csv", encoding = "utf-8", index=False)
```
