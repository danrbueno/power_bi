# Genres dataset

*   This Python script list all the genres from a [public Netflix dataset](https://www.kaggle.com/datasets/victorsoeiro/netflix-tv-shows-and-movies) and creates a new dataset with the list.

```python
import pandas as pd
```

Read the CSV file that contains the titles from Netflix dataset.

```python
titles = pd.read_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/titles.csv")
```

Every row can have more than one genre.

So, it's necessary to separate the genres listed in every row.

This column will be treated as list of strings.

```python
genres = titles.genres
genres
```

Read all the genres listed and create a list with unique values.

```python
single_genres = []

for genre in genres:
  for single_genre in eval(genre):
    single_genres.append(single_genre)
  
unique_genres = list(dict.fromkeys(single_genres))
unique_genres
```

Generates another dataset with the list of genres.



```python
df = pd.DataFrame()
i=1
unique_genres.sort()

for unique_genre in unique_genres:
  row = {'id': str(i),
         'name': unique_genre}
  df = df.append(row, ignore_index=True)
  i = i+1

print(df)

df.to_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/genres.csv", encoding = "utf-8", index=False)
```
