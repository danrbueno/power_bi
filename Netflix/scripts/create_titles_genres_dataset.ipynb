# Titles and genres dataset

*   This Python script list all the titles from a [public Netflix dataset](https://www.kaggle.com/datasets/victorsoeiro/netflix-tv-shows-and-movies) and creates a new dataset with each title and genre in a separated row.

```python
import pandas as pd
```

Read the CSV file that contains the titles from Netflix dataset.

```python
titles = pd.read_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/titles.csv")
titles
```

Read the CSV file that contains the genres from Netflix dataset.

```python
unique_genres = pd.read_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/genres.csv", index_col="name")
unique_genres
```

Every row can have more than one genre.

So, it's necessary to separate the genres listed in every row.

This column will be treated as list of strings.

```python
genres = titles.genres
title = titles.title
genres
```

Read all the genres listed and create a new dataset relationing all titles and genres.

```python
df = pd.DataFrame()
#print(unique_genres.where(unique_genres.genre == 'documentation').head(1).id)
for title in titles.itertuples():
  for genre in eval(title.genres):
      row = {'title_id': title.id,
            'genre_id': str(unique_genres.loc[genre].id)}
      df = df.append(row, ignore_index=True)
  
print(df)
df.to_csv("/content/drive/MyDrive/Portifolio/Netflix/datasets/titles_genres.csv", encoding = "utf-8", index=False)
```
